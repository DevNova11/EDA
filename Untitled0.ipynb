```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
# from statsmodels.datasets import airquality # This line caused an ImportError and is not used.

# Loading the AirPassengers dataset (commonly used time series data)
# This is a common way to get the AirPassengers data in Python
# from statsmodels.datasets import get_rdataset
# data = get_rdataset("AirPassengers", "datasets")
# air_df = data.data
# air_df['Month'] = pd.to_datetime(air_df['time'].apply(lambda x: f'{int(x)}-{int((x % 1) * 12) + 1:02d}-01'))
# air_df = air_df.rename(columns={'value': '#Passengers'})
# air_df = air_df[['Month', '#Passengers']]

# Alternative: Manually create a similar air_df for demonstration if get_rdataset causes issues or is not desired
# For this example, let's create a dummy DataFrame that resembles AirPassengers
# to avoid dependency on R datasets or to ensure reproducibility in Colab easily.

# Let's create a dummy air_df for demonstration purposes that mimics the structure.
# In a real scenario, you would load your actual dataset.
# Here, we'll create a DataFrame similar to what AirPassengers would be.
months = pd.date_range(start='1949-01-01', periods=144, freq='MS')
passengers = np.linspace(100, 600, 144) + np.random.normal(0, 50, 144) # Dummy passenger numbers
air_df = pd.DataFrame({'Month': months, '#Passengers': passengers})
air_df['#Passengers'] = air_df['#Passengers'].astype(int) # Ensure passenger count is integer



# ==========================================
# 1. SPLITTING & CONCATENATION (Using AirPassengers)
# ==========================================
# Split the dataset into two halves to demonstrate vertical concatenation
half1 = air_df.iloc[:18].copy()
half2 = air_df.iloc[18:34].copy()

air_reconstructed = pd.concat([half1, half2], ignore_index=True)
print("\nReconstructed Dataset via Concatenation (First 5 rows):")
print(air_reconstructed.head())


# ==========================================
# 2. MERGING DATAFRAMES
# ==========================================
# Create a dummy lookup table for regional events or notes based on specific dates
events_df = pd.DataFrame({
    'Month': pd.to_datetime(['1949-01-01', '1949-06-01', '1950-01-01']),
    'Event': ['Post-War Peak', 'Summer Rush', 'New Year Travel']
})

# Left Join to keep all air passenger records and match events where available
merged_air = pd.merge(air_df, events_df, on='Month', how='left')
print("\nMerged Data with Event Info (First 10 rows):")
print(merged_air.head(10))


# ==========================================
# 3. ADDING A NEW COLUMN (Feature Engineering)
# ==========================================
# Calculate the year-over-year differences or simple growth indicators
# Let's extract individual time elements as new columns
air_df['Year'] = air_df['Month'].dt.year
air_df['Month_Num'] = air_df['Month'].dt.month

# Example calculation: Projected revenue assuming an average ticket price of $150
air_df['ProjectedRevenue'] = air_df['#Passengers'] * 150

print("\nDataset with New Generated Columns:")
print(air_df.head())


# ==========================================
# 4. DESCRIPTIVE STATISTICS
# ==========================================
print("\nDescriptive Statistics for AirPassengers:")
print(air_df[['#Passengers', 'ProjectedRevenue']].describe())


# ==========================================
# 5. VALUE COUNTS
# ==========================================
# Find out how many data points exist for each year in your snippet
print("\nData Points Count per Year:")
print(air_df['Year'].value_counts())


# ==========================================
# 6. HIERARCHICAL INDEXING (STACK/UNSTACK)
# ==========================================
# Set a multi-index using Year and Month to pivot your data structure
hierarchical_air = air_df.set_index(['Year', 'Month_Num'])[['#Passengers']]
print("\nHierarchical Indexed Air Data:")
print(hierarchical_air.head(15))

# Unstack to view months as columns grouped by year
unstacked_air = hierarchical_air.unstack(level='Month_Num')
print("\nUnstacked Air Data Matrix (Years x Months):")
print(unstacked_air)
```
