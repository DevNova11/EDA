```python
import pandas as pd

df1Basketball = pd.DataFrame({'PlayerID': [1, 3, 5, 7, 9, 11, 21],
                              'PointsBasketball': [15, 22, 10, 30, 8, 18, 5]})
df2Basketball = pd.DataFrame({'PlayerID': [2, 4, 6, 8, 10, 12, 11, 23],
                              'PointsBasketball': [12, 25, 14, 28, 9, 20, 16, 7]})

df1Soccer = pd.DataFrame({'PlayerID': [1, 2, 13, 15, 17, 19, 21],
                           'GoalsSoccer': [2, 1, 3, 0, 2, 1, 0]})
df2Soccer = pd.DataFrame({'PlayerID': [14, 16, 18, 20, 19, 22, 24],
                           'GoalsSoccer': [1, 0, 2, 3, 1, 0, 1]})

print("Basketball Dataframe 1:")
display(df1Basketball)
print("\nBasketball Dataframe 2:")
display(df2Basketball)
print("\nSoccer Dataframe 1:")
display(df1Soccer)
print("\nSoccer Dataframe 2:")
display(df2Soccer)
```

### Combining Dataframes for Each Sport

First, we'll combine the individual dataframes for each sport into a single dataframe for Basketball and a single dataframe for Soccer, similar to how the `dfSE` and `dfML` were created in your example.

```python
dfBasketball = pd.concat([df1Basketball, df2Basketball], ignore_index=True)
dfSoccer = pd.concat([df1Soccer, df2Soccer], ignore_index=True)

print("Combined Basketball Scores:")
display(dfBasketball)
print("\nCombined Soccer Scores:")
display(dfSoccer)
```

### Option 1: Concatenating with `axis=1` (Side-by-Side Join - Often Not Ideal for Different IDs)

This method concatenates dataframes side-by-side. If the indices don't align perfectly (e.g., if PlayerIDs are not in the same order or counts are different), it can lead to `NaN` values and misaligned data. It's generally not recommended for joining based on a common identifier like `PlayerID` without careful index management.

```python
df_concat_axis1 = pd.concat([dfBasketball, dfSoccer], axis=1)
display(df_concat_axis1)
```

### Option 2: Inner Merge (Players in Both Sports)

An inner merge (or inner join) will give us a dataframe containing only the players who appear in *both* the Basketball and Soccer dataframes. This means we get the list of multi-sport players.

```python
df_inner_merge = dfBasketball.merge(dfSoccer, on='PlayerID', how='inner')
print("Players who played in both Basketball and Soccer (Inner Merge):")
display(df_inner_merge)
```

### Option 3: Left Merge (All Basketball Players + Their Soccer Stats)

A left merge (or left join) will include all players from the 'left' dataframe (Basketball in this case) and their corresponding stats from the 'right' dataframe (Soccer). If a Basketball player did not play Soccer, their `GoalsSoccer` will be `NaN`.

```python
df_left_merge = dfBasketball.merge(dfSoccer, on='PlayerID', how='left')
print("All Basketball Players and their Soccer Stats (Left Merge):")
display(df_left_merge)
```

### Option 4: Right Merge (All Soccer Players + Their Basketball Stats)

A right merge (or right join) is the opposite of a left merge. It includes all players from the 'right' dataframe (Soccer) and their corresponding stats from the 'left' dataframe (Basketball). If a Soccer player did not play Basketball, their `PointsBasketball` will be `NaN`.

```python
df_right_merge = dfBasketball.merge(dfSoccer, on='PlayerID', how='right')
print("All Soccer Players and their Basketball Stats (Right Merge):")
display(df_right_merge)
```

### Answering the Head of Department's Questions (Sports Edition)

Now, let's answer questions about player participation, similar to your original problem statement.

```python
df_all_players = pd.concat([dfBasketball[['PlayerID']], dfSoccer[['PlayerID']]]).drop_duplicates()
total_unique_players = df_all_players['PlayerID'].nunique()
print(f"Total unique players across both sports: {total_unique_players}")
```

```python
players_only_basketball = dfBasketball[~dfBasketball['PlayerID'].isin(dfSoccer['PlayerID'])]['PlayerID'].nunique()
print(f"Players who only appeared for Basketball: {players_only_basketball}")
```

## Another Sports Scenario: Game Statistics

Let's consider a single dataset of general game statistics, similar to your `sales.csv` example. We can perform operations like adding new calculated columns, counting occurrences, and getting descriptive statistics.

```python
game_stats_data = {
    'GameID': [101, 101, 102, 102, 103, 103, 104, 104, 105, 105],
    'PlayerID': [1, 2, 3, 4, 1, 5, 6, 7, 3, 8],
    'Team': ['Team A', 'Team B', 'Team A', 'Team B', 'Team A', 'Team B', 'Team A', 'Team B', 'Team A', 'Team B'],
    'FieldGoalsMade': [5, 7, 6, 8, 4, 9, 7, 5, 6, 8],
    'ThreePointersMade': [2, 1, 3, 2, 1, 2, 3, 1, 2, 0],
    'FreeThrowsMade': [3, 2, 4, 3, 2, 3, 1, 2, 4, 1],
    'Assists': [4, 3, 5, 2, 6, 4, 3, 5, 2, 4],
    'Rebounds': [8, 6, 7, 5, 9, 8, 6, 7, 5, 9]
}
game_stats_df = pd.DataFrame(game_stats_data)

print("Original Game Stats Dataframe:")
display(game_stats_df.head(10))
```

```python
game_stats_df['TotalPoints'] = (game_stats_df['FieldGoalsMade'] * 2) + \
                                (game_stats_df['ThreePointersMade'] * 3) + \
                                (game_stats_df['FreeThrowsMade'] * 1)

print("Game Stats Dataframe with TotalPoints:")
display(game_stats_df.head(10))
```

```python
print("Team counts:")
display(game_stats_df['Team'].value_counts())
```

```python
print("Descriptive statistics for Game Stats:")
display(game_stats_df.describe())
```
