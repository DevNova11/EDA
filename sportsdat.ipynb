```python
import pandas as pd

bb_g1 = pd.DataFrame({'PlayerID': [1, 3, 5, 7, 9, 11, 21],
                              'PointsBasketball': [15, 22, 10, 30, 8, 18, 5]})
bb_g2 = pd.DataFrame({'PlayerID': [2, 4, 6, 8, 10, 12, 11, 23],
                              'PointsBasketball': [12, 25, 14, 28, 9, 20, 16, 7]})

soc_g1 = pd.DataFrame({'PlayerID': [1, 2, 13, 15, 17, 19, 21],
                           'GoalsSoccer': [2, 1, 3, 0, 2, 1, 0]})
soc_g2 = pd.DataFrame({'PlayerID': [14, 16, 18, 20, 19, 22, 24],
                           'GoalsSoccer': [1, 0, 2, 3, 1, 0, 1]})

print("Basketball Dataframe 1:")
display(bb_g1)
print("\nBasketball Dataframe 2:")
display(bb_g2)
print("\nSoccer Dataframe 1:")
display(soc_g1)
print("\nSoccer Dataframe 2:")
display(soc_g2)
```

### Combining Dataframes for Each Sport

First, we'll combine the individual dataframes for each sport into a single dataframe for Basketball and a single dataframe for Soccer, similar to how the `dfSE` and `dfML` were created in your example.

```python
bb_all_scores = pd.concat([bb_g1, bb_g2], ignore_index=True)
soc_all_goals = pd.concat([soc_g1, soc_g2], ignore_index=True)

print("Combined Basketball Scores:")
display(bb_all_scores)
print("\nCombined Soccer Scores:")
display(soc_all_goals)
```

### Option 1: Concatenating with `axis=1` (Side-by-Side Join - Often Not Ideal for Different IDs)



```python
df_concat_bad = pd.concat([bb_all_scores, soc_all_goals], axis=1)
display(df_concat_bad)
```

### Option 2: Inner Merge (Players in Both Sports)


```python
df_dual_sport = bb_all_scores.merge(soc_all_goals, on='PlayerID', how='inner')
print("Players who played in both Basketball and Soccer (Inner Merge):")
display(df_dual_sport)
```

### Option 3: Left Merge (All Basketball Players + Their Soccer Stats)



```python
df_bb_plus_soc = bb_all_scores.merge(soc_all_goals, on='PlayerID', how='left')
print("All Basketball Players and their Soccer Stats (Left Merge):")
display(df_bb_plus_soc)
```

### Option 4: Right Merge (All Soccer Players + Their Basketball Stats)



```python
df_soc_plus_bb = bb_all_scores.merge(soc_all_goals, on='PlayerID', how='right')
print("All Soccer Players and their Basketball Stats (Right Merge):")
display(df_soc_plus_bb)
```

### Answering the Head of Department's Questions (Sports Edition)

Now, let's answer questions about player participation, similar to your original problem statement.

```python
all_unique_players_df = pd.concat([bb_all_scores[['PlayerID']], soc_all_goals[['PlayerID']]]).drop_duplicates()
count_all_unique_players = all_unique_players_df['PlayerID'].nunique()
print(f"Total unique players across both sports: {count_all_unique_players}")
```

```python
count_only_bb_players = bb_all_scores[~bb_all_scores['PlayerID'].isin(soc_all_goals['PlayerID'])]['PlayerID'].nunique()
print(f"Players who only appeared for Basketball: {count_only_bb_players}")
```

```python
count_only_soc_players = soc_all_goals[~soc_all_goals['PlayerID'].isin(bb_all_scores['PlayerID'])]['PlayerID'].nunique()
print(f"Players who only appeared for Soccer: {count_only_soc_players}")
```

## Another Sports Scenario: Game Statistics



```python
game_data_dict = {
    'GameID': [101, 101, 102, 102, 103, 103, 104, 104, 105, 105],
    'PlayerID': [1, 2, 3, 4, 1, 5, 6, 7, 3, 8],
    'Team': ['Team A', 'Team B', 'Team A', 'Team B', 'Team A', 'Team B', 'Team A', 'Team B', 'Team A', 'Team B'],
    'FieldGoalsMade': [5, 7, 6, 8, 4, 9, 7, 5, 6, 8],
    'ThreePointersMade': [2, 1, 3, 2, 1, 2, 3, 1, 2, 0],
    'FreeThrowsMade': [3, 2, 4, 3, 2, 3, 1, 2, 4, 1],
    'Assists': [4, 3, 5, 2, 6, 4, 3, 5, 2, 4],
    'Rebounds': [8, 6, 7, 5, 9, 8, 6, 7, 5, 9]
}
df_game_stats = pd.DataFrame(game_data_dict)

print("Original Game Stats Dataframe:")
display(df_game_stats.head(10))
```

```python
df_game_stats['TotalPoints'] = (df_game_stats['FieldGoalsMade'] * 2) + \
                                (df_game_stats['ThreePointersMade'] * 3) + \
                                (df_game_stats['FreeThrowsMade'] * 1)

print("Game Stats Dataframe with TotalPoints:")
display(df_game_stats.head(10))
```

```python
print("Team counts:")
display(df_game_stats['Team'].value_counts())
```

```python
print("Descriptive statistics for Game Stats:")
display(df_game_stats.describe())
```
